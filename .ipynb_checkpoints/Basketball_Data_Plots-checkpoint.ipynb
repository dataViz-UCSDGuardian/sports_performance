```python
%matplotlib inline
from pandas import DataFrame, read_excel
import numpy as np
import matplotlib.pyplot as mpl

# extract win percentage row from Basketball Data table
dfM = read_excel( "Basketball Data.xlsx", sheetname = 0).as_matrix()
dfMP = dfM[24]
dfW = read_excel( "Basketball Data.xlsx", sheetname = 1).as_matrix()
dfWP = dfW[23]
```

```python
#plot men's winning percentage
plt = mpl.plot(dfMP, color = 'blue', marker = '.')
#plot women's winning percentage
plt = mpl.plot(dfWP, color = 'red', marker = '.')
#edit x axis 
plt = mpl.xticks(np.arange(11), np.arange(2005,2016, 1))
```
